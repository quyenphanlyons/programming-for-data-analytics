```python
import requests
import json

# Download the official JSON file from GOV.UK and store in table 'bh'

url = "https://www.gov.uk/bank-holidays.json"
response = requests.get(url)
bh = response.json()
print(bh)
```

```python
# Create new table by electing only 'northern-ireland' data
ni_bh = bh["northern-ireland"]["events"]
print(ni_bh)
```

```python
# Print each holiday with its date
for event in ni_bh:
    print(f"  • {event['date']} – {event['title']}")

print("\nDone!")
```

```python
import json

filename = "bank-holidays.json"

with open(filename,"r") as fp:
    bh = json.load(fp)
    ni_bh = bh["northern-ireland"]["events"]
print(ni_bh)
```
