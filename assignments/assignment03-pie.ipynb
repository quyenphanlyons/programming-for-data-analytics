```python
import pandas as pd
filename = "people-1000.csv"

# Read in the csv file
df= pd.read_csv(filename)
df.head()
```

```python
# Create a new variable that store the domains
# Split() divides a string into a list based on a specified delimiter, in this case is "@". It divides the email address into 2 parts: before and after '@'
splitemail = df["Email"].str.split('@')

# Only get the domain from the list
df["Domain"]=splitemail.str[1]
df.head()
```

```python
# Pie chart

# Count the number of individuals per domain
count = df['Domain'].value_counts()
print(count)

import matplotlib.pyplot as plt
plt.pie(count)
```
