```python
# Data frames.
import pandas as pd

# Create names for the columns
# column_names = ["Statistic Label","Year","Sex","Criteria for Projection","UNIT","VALUE"]

# Load the CSV into a DataFrame:
br = pd.read_csv("births.csv")

# Visualise tableau
print(br.head())
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,6))
# Create histograms
plt.plot(br["Year"], br["VALUE"], marker='o')
plt.title("Projected Births Over Years (2023-2057)")
plt.xlabel("Year")
plt.ylabel("Number of Births")
```
